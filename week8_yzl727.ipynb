#**Encapsulation**

Encapsulation is one of the fundamental concepts in object-oriented programming (OOP). It describes the idea of wrapping data and the methods that work on data within one unit. This puts restrictions on accessing variables and methods directly and can prevent the accidental modification of data. To prevent accidental change, an object’s variable can only be changed by an object’s method. Those types of variables are known as private variables.


A class is an example of encapsulation as it encapsulates all the data that is member functions, variables, etc.



image.png

Kapsülleme, methotların ve variable’ların erişimlerini kısıtlamak anlamına gelmektedir.

A common feature in object oriented programming languages is that classes can usually hide their attributes from any prospective clients. Hidden attributes are usually called private. In Python this privacy is achieved by adding two underscores   **__**    to the beginning of the attribute name

**Example of Encapsulation**



*   Kapsülleme kullanarak, bir nesnenin iç temsilini dışarıdan gizleyebiliriz. Buna **bilgi gizleme** denir.
*   Ayrıca kapsülleme, değişkenlere ve yöntemlere doğrudan erişimi kısıtlamamıza ve bir sınıf içinde özel veri üyeleri ve yöntemler oluşturarak yanlışlıkla veri değişikliğini önlememize olanak tanır.
*   **Kapsülleme**, sınıf dışından yöntemlere ve değişkenlere erişimi kısıtlamanın bir yoludur. Sınıfla çalışırken ve hassas verilerle uğraşırken, sınıf içinde kullanılan tüm değişkenlere erişim sağlamak iyi bir seçim değildir.
*   Örneğin, bir nesnenin dışından görünmeyen bir özniteliğiniz olduğunu ve bunu okuma veya yazma erişimi sağlayan yöntemlerle bir araya getirdiğinizi varsayalım. Bu durumda, belirli bilgileri gizleyebilir ve nesnenin dahili durumuna erişimi kontrol edebilirsiniz. Kapsülleme, programa bir sınıfın tüm değişkenlerine tam kapsamlı erişim sağlamadan gerekli değişkene erişmemiz için bir yol sunar. **Bu mekanizma, bir nesnenin verilerini diğer nesnelerden korumak için kullanılır.**

```python
class Employee:
  #constructor
  def __init__(self,name, salary, project):
    #data members
    self.name=name
    self.salary=salary
    self.project=project

 #method
  def show(self):
    #erişim sağlama
    print("Name: ", self.name, "Salary: ", self.salary)

 #method
  def work(self):
    print(self.name, "üzerinde çalışıyor şu projenin :", self.project)
```

```python
class Employee:
  #constructor

  def __init__(self,name, salary):
    #private data member
    self.__name=name
    self.__salary=salary

  def show(self):
    print("Name:", self.__name, "Salary:", self.__salary)
```

**Name Mangling to access private members**
(Özel üyelere erişmek için isim yöntemi)

We can directly access private and protected variables from outside of a class through name mangling. The name mangling is created on an identifier by adding two leading underscores and one trailing underscore, like this **_classname__dataMember**, where **classname** is the current class, and data member is the private variable name

```python
class Employee:
  #constructor
  def __init__(self, name, salary):
    #public data member
    self.name=name
    #private data member
    self.__salary=salary
```

```python
emp=Employee("Mert Albay", 10000)
```

```python
emp.name
```

```python
emp.salary
```

```python
emp._Employee__salary
```

```python
print("Kapsullenen data member i name mangling yöntemi ile böyle acarlar: ", emp._Employee__salary)
```

**Getter and Setter in Python Encapsulation**

```python
class BankAccount:
  #contructor
  def __init__(self, name, money, address):
    self.name=name #public
    self.__money=money #private
    self.address=address

  #getter setter method

  def getMoney(self):
    return self.__money

  def setMoney(self, amount):
    self.__money=amount

  def __increase(self):
    self.__money=self.__money+500
```

```python
bank1=BankAccount("Yağmur IŞIK",10000, "Ankara")
```

```python
bank1._BankAccount__increase()
```
