#**Inheritance**

Inheritance is the capability of one class to derive or inherit the properties from another class.

Benefits of inheritance are:

It represents real-world relationships well.

It provides the reusability of a code. We donâ€™t have to write the same code again and again. Also, it allows us to add more features to a class without modifying it.

It is transitive in nature, which means that if class B inherits from another class A, then all the subclasses of B would automatically inherit from class A.

##New Class

```python
#Parent Class
class Company:
  #contructor
  def __init__(self, name):
    self.name=name
    print("Company class is created")

  def section(self):
    print("Company section")

  def admin(self):
    print("Company admin")
```

The super() function is a special function that allows you to call a method from the parent class. In this example, call the __init__() method from Company, which gives an AI instance all the attributes defined in that method.
the name super() comes from a convention of calling the parent class a superclass  and the child class a subclass.  

```python
#Child Class
class AI(Company):
  def __init__(self):
    super().__init__(self)
    print("child class is created which name is AI")

  def machinelearning(self):
    print("Using machine learning")

  def deeplearning(self):
    print("Using deep learning")


```

```python
com=Company("Murat")
```

```python
ai=AI()
```

```python
ai.section()
```

```python
ai.name
```

```python
ai.admin()
```

```python
ai.deeplearning()
```

```python
ai.machinelearning()
```

```python
#Parent Class

class Animal:
  def __init__(self):
    print("animal is created")

  def toString(self):
    print("animal")

  def walk(self):
    print("animal walks")
```

```python
#child class

class Monkey(Animal):
  def __init__(self):
    super().__init__() #use init  of parent class (Animal)
    print("monkey is created")

  def toString(self):
    print("monkey")

  def climb(self):
    print("monkey can climb")

#child class

class Bird(Animal):
  def __init__(self):
    super().__init__()
    print("bird is created")

  def fly(self):
    print("Birds can fly")


```

##New Class  (Multiple Inheritance)

```python
#Python example to show the working of multiple inheritance

class Base1:
  def __init__(self):
    self.str1= "Python"
    print("Base1 Class")
  def method(self):
    print("first method")


class Base2:
  def __init__(self):
    self.str2="OOP"
    print("Base2 Class")
  def method(self):
    print("second method")



class Derived(Base1, Base2):
  def __init__(self):

    #calling construuctors of Base1 and Base2 classes
    Base1.__init__(self)
    Base2.__init__(self)
    print("Derived Class")

  def printStrs(self):
    print(self.str1, self.str2)


```

```python
obj=Derived()
```

```python
obj.printStrs()
```

```python
obj.method()
```

#New Class

```python

class Emp():
  def __init__(self,id,name,Add):
    self.id=id
    self.name=name
    self.Add=Add

class Freelance(Emp):
  def __init__(self, id, name, Add, Emails):
    super().__init__(id,name,Add)
    self.Emails=Emails
```

```python
freelance=Freelance(123,"Murat","Ankara","murat.simsek@ostimteknik.edu.tr")
```

```python
print("The ID is:", freelance.id)
print("The name is:", freelance.name)
print("The Address is:", freelance.Add)
print("The Email is:", freelance.Emails)
```
