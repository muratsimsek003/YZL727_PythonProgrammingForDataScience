# Toplulaştırma ve Gruplama (Aggregation & Grouping)
#############################################

# - count()
# - first()
# - last()
# - mean()
# - median()
# - min()
# - max()
# - std()
# - var()
# - sum()
# - pivot table

```python
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/yzl727/dataset.csv")
df.head()
```

```python
df["Age"].mean()
```

```python
df["Gender"].value_counts()
```

```python
df.groupby("Gender")["Age"].mean()
```

```python
df.groupby("Gender").agg({"Age": "mean"})
```

```python
df.groupby("Gender").agg({"Age": ["mean","sum"]})
```

```python
df.groupby("Gender").agg({"Age": ["mean", "sum"],
                       "EstimatedSalary": "mean"})
```

```python
df.groupby(["Gender", "Geography"]).agg({"Age": "mean",
                       "EstimatedSalary": "mean"})
```

```python
df.groupby(["Gender", "Geography"]).agg({"Age": "mean",
                       "Exited": "count"})
```

#######################
# Pivot table
#######################

pivot table veri içerisindeki kırılımları incelemeye yarar.

.pivot_table(ilk değişken kırılımlarda ne görmek istediğimizi, satırda ne olsun, sütünda ne olsun)

```python
df.pivot_table("Age", "Geography", "Exited")
```

buradaki değerler yaş bilgisinin ortalamalarıdır.

```python
df.pivot_table("Age", "Geography", "Exited", aggfunc="std")  #aggfuc fonksiyonu kesişimi ne yapmak istediğimizi için kullanılır
```

```python
df.pivot_table("Age", "Geography", ["Exited", "HasCrCard"])
```

# Birleştirme (Join) İşlemleri
#############################################

```python
import numpy as np
import pandas as pd
m = np.random.randint(1, 30, size=(5, 3))
df1 = pd.DataFrame(m, columns=["var1", "var2", "var3"])
```

```python
df1.head()
```

```python
df2 = df1 + 99
```

```python
df2.head()
```

iki dataframe i birleştirmek için .concat() komutu kullanılır

```python
pd.concat([df1, df2])
```

burada dümdüz alt alta birleştirdi ve indeksler tekrarladı

```python
pd.concat([df1, df2], ignore_index=True) #indeks devam eder
```


#######################
# Merge ile Birleştirme İşlemleri
#######################

```python
df1 = pd.DataFrame({'employees': ['john', 'dennis', 'mark', 'maria'],
                    'group': ['accounting', 'engineering', 'engineering', 'hr']})
```

```python
df1.head()
```

```python
df2 = pd.DataFrame({'employees': ['mark', 'john', 'dennis', 'maria'],
                    'start_date': [2010, 2009, 2014, 2019]})
```

```python
pd.merge(df1, df2)
```

```python
pd.merge(df1, df2, on="employees")
```

```python
df3 = pd.merge(df1, df2)
```

```python
df4 = pd.DataFrame({'group': ['accounting', 'engineering', 'hr'],
                    'manager': ['Caner', 'Mustafa', 'Berkcan']})
```

```python
df4.head()
```

```python
pd.merge(df3, df4)
```
