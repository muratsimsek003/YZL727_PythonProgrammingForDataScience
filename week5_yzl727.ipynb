```python
class MyNum:
  def __init__(self, value):
    try:
      value=int(value)
    except ValueError:
      value=0
    self.value=value
  
  def increment(self):
    self.value=self.value+1
    print(self.value)
```

```python
number=MyNum(10)
```

```python
number.increment()
```

```python
number2=MyNum("Murat")
```

```python
number2.increment()
```

```python
class Araba:
  def __init__(self,manufacturer, model, year):
    self.manufacturer=manufacturer
    self.model= model
    self.year=year

  def welcome(self):
    print("Araba classna  hogeldiniz")

  def getInformation(self):
    long_name=str(self.year)+ " "+self.manufacturer+" "+self.model
    print(long_name)
```

```python
araba=Araba("mercedes","c200",2021)
```

```python
araba.welcome()
```

```python
araba.getInformation()
```

```python
araba.manufacturer
```

```python
araba.year
```
